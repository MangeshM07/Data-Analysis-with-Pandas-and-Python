## Create a MultiIndex with the set_index Method

```python
bigmac = pd.read_csv("bigmac.csv",parse_dates = ["Date"])
bigmac.head(3)
```

```python
bigmac.set_index("Date")
```

```python
bigmac.nunique()
```

```python
#As a best practice, use the columns with least no. of  distinct values as outer index
bigmac.set_index(keys = ["Date","Country"])
bigmac.set_index(keys = ["Country","Date"])
bigmac.set_index(keys = ["Date","Country"],inplace=True)
bigmac.head(3)
```

## Change Index Level names with the set_names() Method

```python
bigmac = pd.read_csv("bigmac.csv",parse_dates = ["Date"],index_col = ["Date","Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.index.set_names(["Day","Location"],inplace=True)
```

```python
bigmac.head()
```

```python
bigmac.index.set_names("Date",level=0)
bigmac.index.set_names("Date",level="Day",inplace = True)
```

```python
bigmac.head()
```

## The sort_index Method on a MultiIndex DataFrame

```python
bigmac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col = ["Date","Country"])
bigmac.head(3)
```

```python
bigmac.sort_index()
```

```python
bigmac.sort_index(ascending = True)
bigmac.sort_index(ascending = False)
```

```python
bigmac.sort_index(ascending = [True,False])
bigmac.sort_index(ascending = [False,True],inplace=True)
bigmac.head()
```

```python
bigmac.sort_index(level=0)
bigmac.sort_index(level="Date")

bigmac.sort_index(level=1)
bigmac.sort_index(level="Country")
bigmac.sort_index(level="Country",ascending=False)
```

## Extract Rows from a MultiIndex DataFrame

```python
bigmac = pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col = ["Date","Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.loc["2010-01-01"]
```

```python
bigmac.loc["2010-01-01","Argentina"]
```

```python
bigmac.loc["2010-01-01","Price in US Dollars"]
```

```python
bigmac.loc[("2010-01-01","Argentina")]
```

```python
bigmac.loc[("2010-01-01","Argentina"),"Price in US Dollars"]
```

```python
#Ideal way to retrive data from MultiIndex DataFrame
bigmac.loc[("2010-01-01","Argentina"),["Price in US Dollars","Price in US Dollars"]]
```

```python
bigmac.loc[("2010-01-01")]
```

```python
bigmac.loc[("2010-01-01",)]
```
